```python
import pandas as pd
import numpy as np
import warnings 
warnings.filterwarnings('ignore')
from sklearn.linear_model import LogisticRegression
```

```python
data=pd.read_csv('framingham.csv')
data.head(n=10)
```

```python
col = ['male','age','education','currentSmoker','cigsPerDay','BPMeds','prevalentStroke','prevalentHyp','diabetes','totChol','sysBP','diaBP','BMI','heartRate','glucose','TenYearCHD']
data = pd.read_csv('framingham.csv',names=col,header=1)
data.head()
```

```python
data.tail()
```

```python
data.count()
```

```python
data.describe()
```

```python
feature = ['diabetes','prevalentStroke','prevalentHyp','currentSmoker','age','male']
```

```python
X = data[feature]
Y = data.TenYearCHD
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(X,Y,random_state=0)
```

```python
log_reg = LogisticRegression()
log_reg.fit(X_train,Y_train)
```

```python
Y_pred = log_reg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score , recall_score
from sklearn.metrics import confusion_matrix
```

```python
accuracy_score(Y_test , Y_pred)
```
